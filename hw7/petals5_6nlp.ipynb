```python
from datasets import load_dataset

dataset = load_dataset("boolq")
dataset = dataset.shuffle()
train_dataset = dataset["train"][:100]
eval_dataset = dataset["validation"][:100]
```

```python
#prompt generation

tc = 0
fc=0
prompt = ""

for sample,l in zip(train_dataset['question'],train_dataset['answer']):
  if l == True :
    if tc<4:
      tc+=1
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+str(l)+"\n"

  elif l == False :
    if fc < 4:
      fc+=1
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+ str(l) +"\n"
  if fc+tc == 8 and fc == tc:
    break
  
```

```python
print(prompt)
```

```python
#evaluate prompt
truelabels = []
tc = 0
fc=0
questions = [eval_dataset["question"]]

prompt_template = "Question: {}\nAnswer: "
```

```python
print(questions)
```

```python
from petals import DistributedBloomForCausalLM
from transformers import AutoModelForCausalLM, AutoTokenizer

checkpoint = "bigscience/bloomz-petals"

  
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = DistributedBloomForCausalLM.from_pretrained(checkpoint)
# Embeddings & prompts are on your device, BLOOM blocks are distributed across the Internet

```

```python
predlabels=[]
for question in questions[0]:
    prompt = prompt_template.format(question)
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_new_tokens=100)
    generated_text = tokenizer.decode(outputs[0])
    answer = generated_text.split("Answer:")[-1].strip()
    predlabels.append(answer)
    print(f"Question: {question}\nAnswer: {answer}\n")
```
