```python
from datasets import load_dataset

dataset = load_dataset("boolq")
dataset = dataset.shuffle()
train_dataset = dataset["train"][:100]
eval_dataset = dataset["validation"][:100]
```

```python
#prompt generation

tc = 0
fc=0
prompt = ""

for sample,l in zip(train_dataset['question'],train_dataset['answer']):
  if l == True :
    if tc<4:
      tc+=1
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+str(l)+"\n"

  elif l == False :
    if fc < 4:
      fc+=1
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+ str(l) +"\n"
  if fc+tc == 8 and fc == tc:
    break
  
```

```python
print(prompt)
```

```python
#evaluate prompt
truelabels = []
tc = 0
fc=0
prompt += "\nAnswer the following questions with 'True or 'False' based on the Question:\n\n"
for sample,l in zip(eval_dataset['question'],eval_dataset['answer']):
  if l == True :
    if tc<15:
      tc+=1
      truelabels.append(str(l))
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+"\n"

  elif l == False :
    if fc < 15:
      fc+=1
      truelabels.append(str(l))
      prompt+='Question: '+str(sample)+"\t"+"Answer: "+"\n"
  if fc+tc == 30 and fc == tc:
    break
```

```python
print(prompt)
```

```python
print(truelabels)
```

```python
import os
import openai

openai.api_key = "randomkey"
```

```python
results = []
response = openai.Completion.create(
  model="text-davinci-003",
  prompt=prompt,
  temperature=0.7,
  max_tokens=818,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)



```

```python
answer = response.choices[0].text.strip()
results.append( answer)
```

```python
print(results)
```

```python
predlabels = results[0].split()[1:]
print(predlabels)
score = 0
for i in range(len(predlabels)):
  if predlabels[i] == truelabels[i]:
    score+=1
print(f"Predicted with an accuracy of {score/len(predlabels) * 100}")
```
